**Dictionary**

The dictionary is Pythonâ€™s built-in mapping type. Dictionaries map keys to values and these key-value pairs provide a useful way to store data in Python.

Typically used to hold data that are related, such as the information contained in an ID or a user profile, dictionaries are constructed with curly braces on either side { }.

In addition to the curly braces, there are also colons (:) throughout the dictionary.

The words to the left of the colons are the keys. Keys can be made up of any immutable data type. 

**Eample Dictionary**

```python
profile = {'username' : 'malka.ra' , 'followers' : 1479, 'paltform' : 'Instagram'}
```

```python
print(profile)
```

```python
type(profile)
```

```python
profile['username']
```

```python
profile['followers']
```

```python
priority = { 1 : 'healt', 2 : 'family', 3 : 'financial'}
```

```python
print(priority)
```

```python
type(priority)
```

```python
priority[1]
```

```python
priority[2]
```

**Using Methods to Access Element**

In addition using keys to access value, we can also work with some buillt-in methods

*   dict.key() isolates keys
*   dict.value() isolates values
*   dict.items() return item in a list format of (key,value) tuple pair





```python
profile = {'username' : 'malka.ra' , 'followers' : 1479, 'paltform' : 'Instagram'}
```

To return the keys, we would use the **dict.keys()** method. In our example, that would use the variable name and be profile.keys().

> Indented block



```python
print(profile.keys())
```

Similarly, we can use the **dict.values() **method to query the values in the sammy dictionary, which would be constructed as profile.values(). 

```python
print(profile.values())
```

If we are interested in all of the items in a dictionary, we can access them with the **items()** method:

```python
print(profile.items())
```

**Modifiying Dictionary**

Dictionaries are a mutable data structure, so you are able to modify them. In this section, weâ€™ll go over adding and deleting dictionary elements.

**adding and Changing Dictionary Element**

```python
profile = {'username' : 'malka.ra' , 'followers' : 1479, 'paltform' : 'Instagram'}
```

```python
print(profile)
```

```python
profile['Country'] = 'Indonesia'
```

```python
print(profile)
```

We see now that the dictionary has been updated with the 'Country': 'Indonesia' key-value pair. Because dictionaries may be unordered, this pair may occur anywhere in the dictionary output. If we use the usernames dictionary later in our program file, it will include the additional key-value pair.

Additionally, this syntax can be used for **modifying** the value assigned to a key. In this case, weâ€™ll reference an existing key and pass a different value to it.

```python
profile = {'username' : 'malka.ra' , 'followers' : 1479, 'paltform' : 'Instagram'}
```

```python
print(profile)
```

```python
profile['followers'] = 2000
```

```python
print(profile)
```

**Deleting Dictionary Elements**

Just as you can add key-value pairs and change values within the dictionary data type, you can also delete items within a dictionary.

```python
profile = {'username' : 'malka.ra' , 'followers' : 1479, 'paltform' : 'Instagram'}
```

```python
print(profile)
```

```python
del profile['paltform']
```

```python
print(profile)
```

If we would like to clear a dictionary of all of its values, we can do so with the** dict.clear()** method. This will keep a given dictionary in case we need to use it later in the program, but it will no longer contain any items.

```python
profile = {'username' : 'malka.ra' , 'followers' : 1479, 'paltform' : 'Instagram'}
```

```python
print(profile)
```

```python
profile.clear()
```

```python
print(profile)
```
